```python
import pandas as pd

# Paste the path you copied from your Google Drive here
file_path = '/content/drive/MyDrive/Online Retail.xlsx' # This is an example path, yours might differ slightly

# Load the data
df = pd.read_excel(file_path)

# Verify it loaded correctly
df.head()
```

```python
# 1. Drop rows where CustomerID is missing
df.dropna(subset=['CustomerID'], inplace=True)

# 2. Convert CustomerID to an integer, then a string, to remove decimals
df['CustomerID'] = df['CustomerID'].astype(int).astype(str)

# 3. Ensure InvoiceDate is a datetime object for calculations
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# 4. Remove returned items (where Quantity is negative)
df = df[df['Quantity'] > 0]

# 5. Create the TotalPrice column (Feature Engineering)
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# --- Verification ---
# Display the info to confirm our cleaning steps worked
print("Data after cleaning:")
df.info()

# Display the first few rows to see the new TotalPrice column
df.head()
```

```python
import datetime as dt

# 1. Set a snapshot_date for Recency calculation
# We'll use the day after the last transaction in the dataset.
snapshot_date = df['InvoiceDate'].max() + dt.timedelta(days=1)

# 2. Group data by each customer and calculate RFM values
rfm_df = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda date: (snapshot_date - date.max()).days, # Recency
    'InvoiceNo': 'nunique',                                        # Frequency
    'TotalPrice': 'sum'                                            # Monetary
})

# 3. Rename the columns to be more descriptive
rfm_df.rename(columns={'InvoiceDate': 'Recency',
                       'InvoiceNo': 'Frequency',
                       'TotalPrice': 'Monetary'}, inplace=True)

# --- Verification ---
# Display the first few rows of your new RFM table
rfm_df.head()
```

```python
# 1. Create labels for Recency, Frequency, and Monetary scores
# For Recency, a lower number of days is better, so we reverse the labels.
r_labels = range(5, 0, -1)
f_labels = range(1, 6)
m_labels = range(1, 6)

# 2. Assign scores from 1-5 based on which quintile the value falls into
rfm_df['R_Score'] = pd.qcut(rfm_df['Recency'], q=5, labels=r_labels)
rfm_df['F_Score'] = pd.qcut(rfm_df['Frequency'].rank(method='first'), q=5, labels=f_labels)
rfm_df['M_Score'] = pd.qcut(rfm_df['Monetary'], q=5, labels=m_labels)

# 3. Combine the scores into a single RFM_Score
rfm_df['RFM_Score'] = rfm_df['R_Score'].astype(str) + rfm_df['F_Score'].astype(str) + rfm_df['M_Score'].astype(str)

# 4. Define segments based on the combined scores using a dictionary and regex
segment_map = {
    r'[4-5][4-5][4-5]': 'Champions',          # High R, F, M
    r'[3-5][3-5][1-3]': 'Loyal Customers',    # Good R & F, low M
    r'[4-5][1-2][1-5]': 'New Customers',      # High R, low F
    r'[2-3][1-3][1-3]': 'Potential Loyalists',# Average R & F, low M
    r'[1-2][3-5][3-5]': 'At-Risk Customers',  # Low R, high F & M
    r'[1-2][1-2][1-2]': 'Hibernating'         # Low R, F, M
}

# 5. Create a new 'Segment' column
rfm_df['Segment'] = rfm_df['RFM_Score'].replace(segment_map, regex=True)

# --- Verification ---
# Display the first few rows to see your new scores and segments
rfm_df.head()
```

```python
# Group by segment and calculate the mean R, F, M and the count of customers
segment_summary = rfm_df.groupby('Segment').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': ['mean', 'count']
}).round(1)

# Display the summary table
print("Customer Segment Summary:")
print(segment_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a bar plot of customer counts by segment
plt.figure(figsize=(12, 6))
sns.countplot(y=rfm_df['Segment'], order=rfm_df['Segment'].value_counts().index)
plt.title('Number of Customers by Segment')
plt.xlabel('Customer Count')
plt.ylabel('Segment')
plt.show()
```

```python
# Identify segments that are still numerical (i.e., they were not matched by the regex)
unsegmented_mask = rfm_df['Segment'].str.match(r'^\d+$')

# Relabel these unsegmented customers as 'Developing'
rfm_df.loc[unsegmented_mask, 'Segment'] = 'Developing'

# --- Verification ---
# Now, re-run your segment summary to see the cleaned-up list
segment_summary_final = rfm_df.groupby('Segment').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': ['mean', 'count']
}).round(1)

print("Final Customer Segment Summary:")
print(segment_summary_final)
```
